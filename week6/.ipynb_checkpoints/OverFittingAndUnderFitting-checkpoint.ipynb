# 过拟合，欠拟合
## 如何解决过拟合问题？
  原因：模型维度过高  
  使用的属性过多  
  解决方法：简化模型结构（使用低阶模型）  
  数据预处理（数据PCA处理）  
  模型训练时候增加正则项  
## 增加正则化
  
  这是线性回归的损失函数
  
  这是正则化处理后的函数，可以理解为损失函数 = 损失函数 + 函数，即为了使损失函数不变，函数要尽可能小，而函数又和$\lambda$与$\theta$有关，$\theta$指代函数g($\theta$,x)里面的$\theta$，g($\theta_i$,$x_i$)可以理解为$\theta_1$ + $\theta_2 * x_1$ + $\theta_3 * x_1$ + $\theta_4 * x_1$ + ....  
  $\lambda$取一个很大的值，来控制各个属性数据影响，例如，$\theta$取很大的数，为了式子大小不变，$\sum_{i=1}^n$$\theta_i^2$必须尽可能小
  也就是说，$\theta_1 \theta_2 \theta_3 \theta_4 $...都要很小才行，这样整个式子大小才不变   
  这样一来，函数g($\theta_i$,x)后面很多项都可以忽略了，也实现了平滑的效果  
  
  这是逻辑回归的最小化损失函数  
    
  对其进行正则化处理，道理相同
