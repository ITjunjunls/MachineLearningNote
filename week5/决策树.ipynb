# 决策树 Decision Tree
一种对实例进行分类的树形结构，通过多层判断区分目标所属类别  
优点：计算量小，运算速度快  
易于理解，可查看各属性重要性  


缺点：忽略属性间的相关性  
样本类别分布不均匀时候，会影响模型表现  

不同特征决定不同的决策树，选择特征的方法：ID3 , C4.5 , CART  
ID3利用信息熵原理选择信息增益最大的属性作为分类属性，完成决策树的构造


这里首先计算量分类前的信息熵，一共有两种结果，分别为是和否，这两种情况分别占$\frac{6}{10}$和$\frac{4}{10}$,代入公式得0.971  
然后我们选择一个特征，计算信息增益，选择有兴趣，他可以分出两类，是和否。观察否，他所对应的最终结果都是否，所以他的信息熵为0，很稳定，再看是，对应的最终结果，有是和否，再计算该特征对应的信息熵得0.361，最终的信息增益为0.61，为其他几个特征对应的信息增益的最高，所以优先选这个作为第一个特征

信息增益最高的分别是兴趣，时间，能力，动力。选择兴趣第一次划分，出现五个否并且是最终正确结果，还剩五个需要再划分，选时间来划分，就获得全部正确结果了，减少了特征数量。
