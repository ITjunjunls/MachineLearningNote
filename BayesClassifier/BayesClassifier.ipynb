# BayesClassifier贝叶斯分类器  


先验概率：可以理解为类别，就比如说男生和女生的面部特征，这是两个类别，他们分别有自己的概率，例如男生和女生人数比为2:1，对应的概率为66%和33%，这就是先验概率  
条件概率：可以理解为男生的那些面部特征所对应的概率和女生面部特征对应的概率  
后验概率：指通过特征来判断属于那个类别，属于不同特征的概率

先验概率是指根据以往经验和分析得到的概率,如全概率公式,它往往作为"由因求果"问题中的"因"出现.   
后验概率是指依据得到"结果"信息所计算出的最有可能是那种事件发生,如贝叶斯公式中的,是"执果寻因"问题中的"因".   

第一行的所有属性的联合概率，指的是有很多种组合，例如有三种职业，每种职业都有两种状态，这样就有很多种情况，他们的概率就不好算
